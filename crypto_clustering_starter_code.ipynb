# Clustering Crypto

```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# read in data from csv file and set the index to 0
crypto_df=pd.read_csv('crypto_data.csv', index_col=0)
crypto_df.head(15)


# Load the crypto_data.csv dataset.
# YOUR CODE HERE
```

```python
# Keep all the cryptocurrencies that are being traded.
crypto_df = crypto_df.loc[(crypto_df['IsTrading']==True)]
crypto_df.head(15)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
crypto_df.shape
```

```python
# Remove the "IsTrading" column. 
crypto_df = crypto_df.drop(columns=['IsTrading'])
crypto_df
```

```python
# Remove rows that have at least 1 null value.
crypto_df = crypto_df.dropna()
crypto_df.head(10)
```

```python
# Keep the rows where coins are mined.
crypto_df = crypto_df.loc[(crypto_df['TotalCoinsMined']>0)]
crypto_df.head(15)
```

```python
crypto_df.shape
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
cc_name_df = pd.DataFrame(crypto_df.CoinName)

cc_name_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
crypto_df=crypto_df.drop(columns=['CoinName'])

print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Use get_dummies() to create variables for text features.
crypto_features = pd.get_dummies(crypto_df, columns=['Algorithm', 'ProofType'])
crypto_features
```

```python
# Standardize the data with StandardScaler().

crypto_scaler= StandardScaler().fit_transform(crypto_features)
crypto_scaler[0:10]
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Create a DataFrame with the three principal components.

# Initialize PCA model
pca = PCA(n_components=3)
# Get two principal components for the iris data.
crypto_pca = pca.fit_transform(crypto_scaler)
crypto_pca
```

```python
# Transform PCA data to a DataFrame
pcs_df = pd.DataFrame(
    data=crypto_pca, columns = ['PC 1', 'PC 2', 'PC 3'], index=crypto_df.index)
pcs_df.head()
```

```python

pcs_df.shape
```

```python
# Get the explained variance
pca.explained_variance_ratio_
```

```python
# The PC 1 contains 2.79% of the variance, PC 2 contains 2.14% of the variance and PC 3 contains 2.05% of the variance
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
##creating an empty list to hold inertia value
inertia = []
k = list(range(1, 11))


# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
    
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initialize the K-means model
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)

# Add the predicted class columns

pcs_df.head(10)
```

```python
# print out the predictions
predictions
```

Running K-Means with `k=4`

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
cluster_df = pd.concat([crypto_df, pcs_df], axis=1, join='inner')

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies.
cluster_df['CoinName'] = cc_name_df.CoinName

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
cluster_df['Class'] = predictions
print(cluster_df.shape)
cluster_df.head(10)
```

### Deliverable 4: Visualizing Cryptocurrencies Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
# Plot the 3D-scatter with x="Annual Income", y="Spending Score (1-100)" and z="Age"


fig = px.scatter_3d(
    cluster_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="Class",
    symbol="Class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
